## Create a Series Object from a List
- A `Series` is a single column of ordered homogenous data (data of the same type).
- A `Series` combines the the best features of a list and a dictionary.
- Pandas assigns an identifier to each value. The identifiers make up the **index** of the `Series`.
- If we do not provide a custom index, Pandas defaults to an ascending numeric index.
- The index starts counting from 0.

````python
ice_cream = ["Chocolate", "Vanilla", "Strawberry", "Rum Raisin"]
pd.Series(ice_cream)
````

- A `Series` output includes the data type (`dtype`) of its values.
- `int64` is a 64-bit integer (whole number occupying 64-bits/8 bytes of memory)

````python
lottery_numbers = [4, 8, 15, 16, 23, 42]
pd.Series(lottery_numbers)
````

- A `Series` can contain duplicates.

````python
pd.Series([True, False, False, False, True])
````

## Create a Series Object from a Dictionary
- The `Series` index does not have to be numeric. It can be a variety of data types.
- If we pass a dictionary to the `Series` constructor, Pandas will use the keys as the index labels.
- Pandas will keep track of each element's order (even if the index is non-numeric).

````python
sushi = {"Salmon": "Orange", "Tuna": "Red", "Eel": "Brown"}
pd.Series(sushi)
````

## Parameters and Arguments
- A **parameter** is the name for an expected input to a function/method/class instantiation.
- An **argument** is the concrete value we provide for a parameter during invocation.
- We can pass arguments either sequentially (based on parameter order) or with explicit parameter names written out.

````python
fruits = ["Apple", "Orange", "Plum", "Grape", "Blueberry", "Watermelon"]
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Monday"]
````

- The first parameter for the `Series` constructor is `data`, which represents the values of the `Series`.
- The constructor will provide fallback/default values to the other parameters.
- Use Shift + Tab to see the `Series` documentation and the default values.
- The `Series` values can contain duplicates.

````python
pd.Series(fruits)
pd.Series(data=fruits)
````

- The second parameter for the `Series` constructor is `index`, which holds the index labels.
- Notice that the `Series` index _can_ contain duplicates (although this is not ideal).

````python
pd.Series(fruits, weekdays)
pd.Series(data=fruits, index=weekdays)
````

- Reversing the order of arguments results in a different `Series`.
- `weekdays` becomes the argument for `data`, and `fruits` becomes the argument for `index`.

````python
pd.Series(weekdays, fruits)
````

- If using keyword arguments, the order does not matter.
- Python matches the parameter to the argument by its name, not by its order.
- Keyword arguments also provide context on what the parameter represents.

````python
pd.Series(data=fruits, index=weekdays)
pd.Series(index=weekdays, data=fruits)
````

- You can mix-and-match positional arguments with keyword arguments.
- Use keyword arguments to skip required arguments.
- The `name` parameter assigns a name to the `Series`.

````python
pd.Series(fruits, index=weekdays, name="Diet")
````

- A `Series` can be empty. Pandas will assume an `object` data type.

````python
pd.Series()
````

## Copy-on-Write and the copy Method
- In earlier versions of Pandas (`< 3.0`), targeting data could return a view.
- A view was a window/alternate perspective into the *same* underlying data.
- Changes to the view would mutate the original object that the view was referencing.
- Pandas 3.0 follows a **Copy-on-Write** principle.
    1. Any object derived from another behaves as an independent copy.
    2. When code mutates the underlying data, Pandas will create a copy to avoid unexpected side effects.
- A **copy** is a duplicate/clone of an object.

- Let's assign the imported `DataFrame` to a `pokemon_df` variable.
- We'll then use it as the basis of creating a `Series` assigned to `pokemon_series`.
- `pokemon_series` share the same data under the hood as the `DataFrame`.
- Pandas has no reason to copy the data yet (and use more memory).

````python
pd.__version__
````

````python
pokemon_df = pd.read_csv("pokemon.csv", usecols=["Name"])
pokemon_series = pokemon_df.squeeze("columns")
````

- Let's use the `iloc` accesor to update the `Series` at index position 0.
- Pandas sees the mutation of the underlying data. It creates a copy of the data for the `Series`.

````python
pokemon_series.iloc[0] = "Borisaur"
````

````python
pokemon_series.head()
````

- The original `pokemon_df` `DataFrame` is **not** mutated.
- The Copy-on-Write principle in Pandas 3 force a copy of the `Series` data when we mutate it (`iloc`).
- `pokemon_series` _was_ relying on shared data before `iloc` but now has its own copy of the data.

````python
pokemon_df
````

- In previous version of Pandas, we would see different behavior.
- Mutating the `Series` would mutate the `DataFrame` as well.
- In previous versions, invoke the `copy` method explicitly to force a copy of the `Series` data.

````python
pokemon_df = pd.read_csv("pokemon.csv", usecols=["Name"])
pokemon_series = pokemon_df.squeeze("columns").copy()
````

## Pandas Aligns by Index Label
- We can perform mathematical operations between multiple `Series`.
- Pandas will align values by index labels, not by position.
- The resulting index will be the union (combination) of all index labels from the two `Series`.
- `NaN` is Pandas' designation for a missing value.
- Any label present in one `Series` but missing in the other produces `NaN` in the result. Pandas cannot add "unknown" to a valid numeric value.

````python
revenue_ny = pd.Series([100, 200, 300], index=["Mon", "Tue", "Wed"])
revenue_ny
````

````python
revenue_la = pd.Series([50, 60, 70], index=["Tue", "Wed", "Thu"])
revenue_la
````

- Let's find the total business revenue per day/index label.
- The `Tue` and `Wed` labels are found in both `Series.`
- The `Mon` label is only found in `revenue_ny`.
- The `Thu` label is only found in `revenue_la`.

````python
revenue_ny + revenue_la
revenue_ny - revenue_la
````

- The `add` method supports a `fill_value` parameter to provide a fallback value for missing labels.
- Pandas will assign the present label value with the `fill_value`.

````python
revenue_ny.add(revenue_la, fill_value=0)
````

## The map Method
