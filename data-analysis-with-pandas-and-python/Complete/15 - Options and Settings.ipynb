# Options and Settings

```python
import pandas as pd
```

## Changing Options with Attributes
- The `pd.options` module configures Pandas's settings.
- The `display` module deals with the presentation of data.

```python
spotify = pd.read_csv("spotify.csv")
spotify.head()
```

- The `max_rows` attribute sets the threshold/limit where the `DataFrame` rows will be truncated.
- The `min_rows` attribute sets the minimum number of displayed rows when `max_rows` is exceeded.
- Overwrite an attribute's value with an equal sign and the new value.

```python
pd.options.display.max_rows
pd.options.display.min_rows
```

```python
pd.options.display.min_rows = 14
```

```python
spotify
```

- The `max_columns` attribute sets the threshold/limit where the `DataFrame` columns will be truncated.

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_columns = 6
```

```python
spotify.head()
```

## Changing Options with Functions
- The `pd.get_option` function retrieves the current value for a pandas setting.
- The `pd.set_option` function sets a new value for a pandas setting.
- The `pd.describe_option` function provides documentation for a pandas setting.
- The `pd.reset_option` function returns a pandas setting back to its default value.

```python
spotify = pd.read_csv("spotify.csv")
spotify.head(1)
```

```python
pd.options.display.min_rows
```

```python
pd.get_option("display.min_rows")
pd.get_option("display.max_rows")
pd.get_option("display.max_columns")
```

```python
pd.set_option("display.min_rows", 14)
pd.set_option("display.max_columns", 6)
```

```python
pd.describe_option("display.max_rows")
```

```python
pd.reset_option("display.min_rows")
pd.reset_option("display.max_rows")
pd.reset_option("display.max_columns")
```

```python
spotify
```

## The precision Option
- The `display.precision` option sets the number of digits that appears after a decimal point in a floating-point number.
- The default value is 6.

```python
spotify = pd.read_csv("spotify.csv")
spotify.head()
```

```python
pd.options.display.precision
```

```python
pd.get_option("display.precision")
```

```python
pd.options.display.precision = 2
```

```python
pd.set_option("display.precision", 2)
```

```python
spotify.head()
```

```python
spotify["Danceability"].iloc[0]
```

## Temporary Options with option_context
- The `pd.option_context` context manager temporarily changes options within a `with` block.
- Options revert back to their default values when the block ends.
- Pass a dictionary where the keys are the settings and the values are their temporary values.
- Use the `display` function to output the `DataFrame`.

```python
pd.reset_option("display.max_rows")
pd.reset_option("display.max_rows")
pd.reset_option("display.max_columns")
pd.reset_option("display.precision")
```

```python
spotify = pd.read_csv("spotify.csv")
spotify
```

```python
with pd.option_context({"display.precision": 3, "display.max_columns": 4}):
    display(spotify)
```
