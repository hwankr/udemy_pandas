## Methods and Attributes between Series and DataFrames
- A `DataFrame` is a 2-dimensional table consisting of rows and columns.
- Like with a `Series`, Pandas assigns an index position/label to each `DataFrame` row.
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `pd.read_csv` function imports a `DataFrame` by default.
- The `nba.csv` dataset is a collection of basketball players.
- The last row in the dataset is completely empty.

```python
nba = pd.read_csv("nba.csv")
nba
```

- The `DataFrame` and `Series` have common and exclusive methods/attributes.
- Some methods/attributes will return different types of data.

```python
s = pd.Series([1, 2, 3, 4, 5])
s
```

- The `head` method returns rows from the start of the `DataFrame`.
- The `tail` method returns rows from the end of the `DataFrame`.

```python
nba.head()
nba.head(n=5)
nba.head(8)
```

```python
nba.tail()
nba.tail(n=7)
```

- The `index` attribute stores the object that holds the index labels.

```python
s.index
nba.index
```

- The `shape` attribute is a tuple with the dimensions (height x width) of the data structure.

```python
s.shape
nba.shape
```

- The `dtypes` attribute holds the data types.
- The `DataFrame` has multiple columns, so Pandas creates a `Series` where the index labels are the column names and the values are their data types.

```python
s.dtypes
nba.dtypes
```

- The `hasnans` attribute returns `True` if a `Series` contains any missing values.
- The attribute does not exist on a `DataFrame`.

```python
s.hasnans
# nba.hasnans
```

- The `columns` attribute returns the column labels of a `DataFrame`.
- The attribute does not exist on a `Series`.

```python
nba.columns
# s.columns
```

- The `axes` attribute returns a list of the axis indexes.
- The `Series` has only one index, so the list has 1 entry.
- The `DataFrame` has two indexes (rows and columns), so the list has 2 entries.

```python
s.axes
nba.axes
```

- The `info` method returns a summary of the pandas object.
- The output includes the index type, a count of non-null values, the data types, and the memory usage.

```python
s.info()
```

- The `info` method on a `DataFrame` outputs the details _per_ column.

```python
nba.info()
```

## Differences between Shared Methods
- The same method may work differently on a `Series` vs. a `DataFrame`.

```python
s = pd.Series([1, 2, 3, 4, 5])
revenue = pd.read_csv("revenue.csv", index_col="Date")
revenue.head()
```

- The `sum` method adds a `Series`'s values. There is only one direction to add.

```python
s.sum()
s.sum(axis="index")
s.sum(axis=0)
```

```python
s.axes
```

- On a `DataFrame`, the `sum` method defaults to adding the column values.
- The method returns a `Series` where the index labels are the column names and the values are the sum totals.
- The `axis` parameter customizes the direction that we add across. Pass `"index"` or `0` to add values by traversing the rows.

```python
revenue.sum()
revenue.sum(axis="index")
revenue.sum(axis=0)
```

- Pass the `axis` parameter an argument of `"columns"` to calculate the sum of each row's values.
- We are traversing the "columns" in calculating the sum rather than traversing the "index".

```python
revenue.sum(axis="columns")
revenue.sum(axis=1)
```

```python
revenue.axes
```

- Now that we have a `Series` of numbers, we can also calculate total business revenue by invoking `sum`.

```python
revenue.sum(axis=1).sum()

revenue.sum(axis=0).sum()
```

## Drop Rows with Missing Values
- Pandas designates missing values with a `NaN`.

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

- The `dropna` method deletes rows with missing values.
- The method's default behavior is to remove a row if it contains *any* missing values.

```python
nba.dropna()
nba.dropna(how="any")
```

- Pass the `how` parameter an argument of "all" to remove rows where all the values are `NaN`.
- This solution effectively removes the final `DataFrame` rows which holds only missing values.

```python
nba.dropna(how="all")
```

- The `subset` parameter limits the columns that pandas will use to drop rows with missing values.
- The next example drop rows with a missing value in either the `College` or `Salary` column (or both).

```python
nba.dropna(subset=["College"])
nba.dropna(subset=["College", "Salary"])
```

- The `thresh` parameter keeps rows with a minimum number of present values.
- `thresh` is useful when some missing data is acceptable and you want to drop rows that are *mostly* empty.
- The next example passes an argument of 5, which keeps rows with at least 5 non-null values (out of 7 columns).

```python
nba.dropna(thresh=5)
```

## Fill in Missing Values with the fillna Method
- The `fillna` method replaces missing `NaN` values with its argument.
- The `fillna` method is available on both `DataFrame` and `Series` objects.
- The `fillna` method returns a new object with missing values replaced.  The original is not mutated.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba.tail()
```

```python
nba.fillna(0)
```

- Pass `fillna` a **dictionary** to specify replacement values _per_ column.
- The dictionary keys will be the column names, and the dictionary values are the replacement values.

```python
nba.fillna({"Salary": 0, "College": "Unknown", "Position": "N/A"})
```

- The next example replaces `NaN` values with 0 in the `Salary` column.

```python
nba["Salary"] = nba["Salary"].fillna(0)
```

```python
nba
```

```python
nba["College"] = nba["College"].fillna(value="Unknown")
```

```python
nba
```

## Fill Missing Values with a Forward Fill or Back Fill
- Forward fills and backward fills are two alternate stategies for replacing missing values.
- Fills are ideal when the previous or next observation is a reasonable estimate for a missing value.
- The `temperatures.csv` file has missing values in the `High Temp` column.
- Example: When measuring temperatures, it's reasonable to guess that a day's temperature will be similar to the one on the previous or next day.

```python
temperatures = pd.read_csv("temperatures.csv")
temperatures.head()
```

- The `ffill` method **forward-fills** missing values.
- Pandas replaces each `NaN` with the last valid value above it.
- For example, index 2's value of 33.0 replaces index 3's missing `NaN` value.

```python
temperatures["High Temp"].ffill()
```

- The `bfill` method **backward-fills** missing values.
- Pandas replaces each `NaN` with the next valid value below it.
- For example, index 4's value of 38.0 replaces index 3's missing `NaN` value.

```python
temperatures["High Temp"].bfill()
```

- The `limit` parameter places a boundary on how many consecutive `NaN` values to fill.
- The values at row indexes 9, 10, and 11 are all `NaN`.
- An argument of 1 replaces 1 `NaN` value in the sequence (index 9) while keeping the other `NaN`s (index 10, 11).

```python
temperatures["High Temp"].ffill(limit=2)
```

## Convert Data Types with the astype Method I
- The `astype` method converts a `Series's` values to a specified type.
- Developers sometimes use the word "cast" to describe type conversion (i.e., "we cast this number from int to float').
- Pandas cannot convert `NaN` values to numeric types, so we need to eliminate/replace them before we perform the conversion.

```python
nba = pd.read_csv("nba.csv").dropna(how="all").fillna({"Salary": 0})
nba.tail()
```

- Pass in the specified type as either a string or the core Python data type.
- The `dtypes` attribute returns a `Series` with the `DataFrame`'s columns and their types.

```python
nba.dtypes
```

```python
nba["Salary"].astype("int")
nba["Salary"].astype(int)
```

- We can then use the assignment operator (`=`) to replace the column.

```python
nba["Salary"] = nba["Salary"].astype(int)
```

```python
nba.dtypes
```

- The second option is to invoke `astype` on the `DataFrame` and pass in a dictionary.
- The keys should be the column names, and the values should be the desired data types.

```python
nba.astype({"Salary": int})
```

- A 64-bit integer (`int64`) can represent values from -9,223,372,036,854,775,808 (quadrillion) to 9,223,372,036,854,775,807.
- A 32-bit integer (`int32`) can represent values from -2,147,483,648 to 2,147,483,647.
- A 16-bit integer (`int16`) can represent values from -32,768 to 32,767.
- A 32-bit integer should be large enough to model every salary in the dataset.
- The `memory_usage` method returns the amount of bytes that each columnn's data occupies in memory.

```python
nba.memory_usage()
nba.info()
```

- Pass the `astype` method an instance of a data type at the top-level of pandas.

```python
nba["Salary"] = nba["Salary"].astype(pd.Int32Dtype())
```

- The data type will update from `int64` to `Int32`.
- Notice the difference in casing. The two are actually different types from different libraries.
- `int64` comes from `numpy` and `Int32` comes from Pandas.
- The Pandas integers support missing values, while the NumPy integers do not support them.

```python
nba.dtypes
```

```python
nba.memory_usage()
nba.info()
```

```python
nba["Weight"] = nba["Weight"].astype(pd.Int16Dtype())
```

```python
nba.info()
```

## Sort a DataFrame with the sort_values Method I
- The `sort_values` method sorts a `DataFrame` by the values in one or more columns.
- If sorting by a single column, pass a string with its name.
- The default sort order is ascending (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba.tail()
```

```python
nba.sort_values("Name")
nba.sort_values(by="Name")
nba.sort_values(by="Name", ascending=True)
nba.sort_values(by="Name", ascending=False)
```

```python
nba.sort_values("Salary")
nba.sort_values("Salary", ascending=True)
nba.sort_values("Salary", ascending=False)
nba.sort_values("Salary", na_position="last")
nba.sort_values("Salary", na_position="first")
nba.sort_values("Salary", na_position="first", ascending=False)
nba.sort_values("Salary", ascending=False, na_position="first")
```

## Sort a DataFrame with the sort_values Method II: Multiple Columns
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba.tail()
```

```python
nba.sort_values(["Team", "Name"])
nba.sort_values(by=["Team", "Name"])
nba.sort_values(by=["Team", "Name"], ascending=True)
nba.sort_values(by=["Team", "Name"], ascending=False)
nba.sort_values(by=["Team", "Name"], ascending=[True, False])
```

```python
nba.sort_values(["Position", "Salary"])
nba.sort_values(by=["Position", "Salary"])
nba.sort_values(by=["Position", "Salary"], ascending=True)
nba.sort_values(by=["Position", "Salary"], ascending=False)

nba.sort_values(by=["Position", "Salary"], ascending=[True, False])
nba.sort_values(by=["Position", "Salary"], ascending=[False, True])
```

## Sort with Ordered Categories
- By default, category values have no ranking.

```python
hotels = pd.read_csv("hotel_bookings.csv")
hotels.head()
```

- The `nunique` methods returns a count of unique values.
- The `unique` method returns a list of all distinct/unique values.

```python
hotels["Room"].nunique()
hotels["Room"].unique()
```

```python
hotels["Room"] = hotels["Room"].astype("category")
```

```python
hotels.sort_values("Room")
```

- An **ordered** categorical assigns a rank to the sequence (e.g., "Standard" < "Deluxe" < "Suite" < "Presidential").
- We'll create a `CategoricalDtype` instance with an ordered list of categories.
- "Standard" comes before "Deluxe", which comes before "Suite".

```python
room_type = pd.CategoricalDtype(
    categories=["Standard", "Deluxe", "Suite", "Presidential"], ordered=True
)
```

```python
hotels["Room"] = hotels["Room"].astype(room_type)
```

- The `sort_values` method will use category order, not alphabetical order.

```python
hotels.sort_values("Room")
```

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each `Series` value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

```python
nba = (
    pd.read_csv("nba.csv")
    .dropna(how="all")
    .fillna({"Salary": 0})
    .astype({"Salary": int})
)
nba.head()
```

```python
nba["Salary"].rank()
nba["Salary"].rank(ascending=True)
nba["Salary"].rank(ascending=False)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False)
```

```python
nba.sort_values("Salary", ascending=False)
```

- The `method` parameter controls how Pandas assigns a rank when multiple values tie for the same rank.

```python
scores = pd.Series([100, 90, 90, 80])
scores
```

- The `"average"` value (default) uses the average of the rankings.

```python
scores.rank(ascending=False, method="average")
```

- The `"min"` value assigns all tied values the lowest rank.

```python
scores.rank(ascending=False, method="min")
```

- The `"max"` value assigns all tied values the highest rank.

```python
scores.rank(ascending=False, method="max")
```

- The `"dense"` value operates like `"min"` but has gaps in the ranking.
- The values of 90 and 90 tie for a rank of 2.
- The value of 100 is assigned a rank fo 3 instead of 4.

```python
scores.rank(ascending=False, method="dense")
```

- The `"first"` value breaks up ties based on the number's position.
- The first occurrence of the duplicate (90) in the `Series` gets the smaller rank.

```python
scores.rank(ascending=False, method="first")
```
