# Input and Output
- The `read_csv` function supports dozens of parameters for handling various file formats.
- All datasets are in the `io_examples` folder.

```python
import pandas as pd
```

- Some datasets may not have a header column. Pandas will treat the first row as the colum names.
- Passing `header=None` creates a numeric index for the columns.
- The `names` parameter can set custom column names.

```python
pd.read_csv("io_examples/sensor_data.csv")
pd.read_csv("io_examples/sensor_data.csv", header=None)
pd.read_csv(
    "io_examples/sensor_data.csv",
    header=None,
    names=["timestamp", "temperature", "humidity"],
)
```

- Datasets may have comments or blank lines before the actual data.
- The `skiprows` parameter skips a specified number of lines at the top of the file.

```python
pd.read_csv("io_examples/quarterly_report.csv", skiprows=3)
```

- By default, pandas recognizes common missing-value markers like `NaN`, `NA`, and empty strings.
- Use `na_values` to define additional values that qualify as missing values.

```python
pd.read_csv("io_examples/customer_data.csv", na_values=["--", "missing"])
```

- The `dtype` parameter sets column types on import.
- The dataset loses the leading zeroes for the `zipcode` column (like `00501`). Conversion later can't help.
- Pass the `dtype` parameter a dictionary with column names as keys and data types as values.

```python
pd.read_csv(
    "io_examples/customer_data.csv", na_values=["--", "missing"], dtype={"zipcode": str}
)
```

- The `nrows` parameter imports only the first `n` rows.
- The parameter can preview a large dataset without loading the complete filev into memory.
- The `index_col` parameter sets a column's values as the index labels.

```python
pd.read_csv("fortune1000.csv", nrows=5)
pd.read_csv("fortune1000.csv", nrows=5, index_col="Rank")
```

## Export DataFrame to CSV File
- The `to_csv` method exports a `DataFrame` to a CSV file.
- Its first argument is the filename.
- By default, pandas will include the index. Set the `index` parameter to False to exclude the index.
- The `columns` parameter limits the exported columns.

```python
url = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
```

```python
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
for column in ["Gender", "Ethnicity", "Child's First Name"]:
    baby_names[column] = baby_names[column].str.strip().str.title()
```

```python
baby_names.head()
```

```python
baby_names.to_csv(
    "baby_names.csv",
    index=False,
    columns=["Year of Birth", "Child's First Name", "Count"],
)
```

## Import Excel Workbook into pandas
- The `read_excel` function reads an Excel file/workbook into a `DataFrame`.

```python
pd.read_excel("Data - Single Worksheet.xlsx")
```

- Use the `sheet_name` parameter if the workbook contains multiple worksheets.
- Pass a single worksheet name or a list of worksheet names/index positions.
- Pandas will store multiple worksheets in a Python dictionary.
- The keys will be the worksheet names, and the values will be the `DataFrame` objects.

```python
pd.read_excel("Data - Multiple Worksheets.xlsx")
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name="Data 1")
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name="Data 2")
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=0)
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=1)
```

```python
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=["Data 1", "Data 2"])
type(pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=["Data 1", "Data 2"]))
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=[0, 1])
```

- Pass the `sheet_name` parameter an argument of `None` to include all worksheets.
- The `keys` method on a dictionary returns a collection of its keys.
- The `values` method on a dictionary returns a collection of its values.

```python
data = pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=None)
data
```

```python
data.keys()
```

```python
data["Data 1"]
data["Data 2"]
```

```python
pd.concat(data.values())
pd.concat(data.values(), ignore_index=True)
```

## Export Excel Workbook from pandas
- The `read_excel` function writes a `DataFrame` to an Excel workbook.
- By default, the worksheet name will be `Sheet1`.
- The `sheet_name` parameter customizes the worksheet name.
- By default, pandas will keep the index as a column. Set `index=False` to exclude the index.
- The `columns` parameter limits the columns that pandas will write.

```python
url = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
males = baby_names[baby_names["Gender"] == "MALE"]
females = baby_names[baby_names["Gender"] == "FEMALE"]
```

```python
females.to_csv("NYC Baby Data.csv")
females.to_excel("NYC Baby Data.xlsx")
females.to_excel("NYC Baby Data.xlsx", index=False)
females.to_excel("NYC Baby Data.xlsx", index=False, sheet_name="Females")
females.to_excel(
    "NYC Baby Data.xlsx",
    index=False,
    sheet_name="Females",
    columns=["Year of Birth", "Child's First Name", "Rank"],
)
```

- The `ExcelWriter` class writes one or more `DataFrame` objects to an Excel file.
- Use a context manager (the `with` keyword) in combination with the `ExcelWriter` object and an assigned variable.
- Invoke the `to_excel` method on every `DataFrame` to include in the Excel workbook and pass in the `ExcelWriter` object as the first argument.
- The `to_excel` method supports `sheet_name`, `index`, and `columns` parameters.

```python
with pd.ExcelWriter("NYC Baby Data.xlsx") as excel_file:
    females.to_excel(
        excel_file,
        sheet_name="Females",
        index=False,
        columns=["Year of Birth", "Child's First Name", "Rank"],
    )
    males.to_excel(excel_file, sheet_name="Males", index=False)
```
