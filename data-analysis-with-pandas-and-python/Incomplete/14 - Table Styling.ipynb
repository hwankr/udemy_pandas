# Table Styling

## The Styler Object
- The `style` attribute on a `DataFrame` returns a `Styler` object.
- A `Styler` renders an HTML table with CSS styles in Jupyter notebooks.
- The `Styler` does not change the underlying data, only its presentation.
- Methods on the `Styler` can customize formatting, colors, highlights, and more.
- The `Styler`'s string representation is the `DataFrame` itself (it's a weird design decision).

- The `style.set_caption` method adds a title above the styled table.
- Methods on the `Styler` return a new `DataFrame`.

- The `style.format` method alters the presentation of values.
- The `precision` parameter sets the number of places after the decimal point.
- The `na_rep` parameter sets a string to display in place of `NaN` values.

- Methods on the `Styler` return the `Styler`, so we can use method chaining.
- We do _not_ need to access `style` again because a `Styler` method does not return the `DataFrame`.

## Hide Index or Column Labels
- The `hide` method hides the row or column index.
- Pass an argument of `axis="columns"` to hide the column headers.

- Pass a `subset` parameter to hide specific rows or columns by label.

## Highlighting Values
- The `highlight_null` method colors the background of cells that contain `NaN`.
- The default color is red. Use the `color` parameter to customize the highlight color.
- Provide the 6-character hexadecimal color code or a color name (like `"gold"`).

- The `highlight_max` and `highlight_min` methods color the highest and lowest values.
- The `axis` parameter controls direction. `0` is max/min per-column, `1` is max/min per-row, or `None` for the entire table.
- The `color` parameter sets the highlight color.

- The `highlight_between` method colors values that fall within a given range.
- The `left` and `right` parameters set the lower and upper bounds. Both are inclusive.

## Using Color Gradients

- The `background_gradient` method creates a heatmap across cells.
- Greater values will have a deeper shade of color.
- The `cmap` parameter accepts any Matplotlib colormap (e.g., `"YlOrRd"`, `"Blues"`, `"viridis"`).
- The `axis` parameter controls direction: `0` per-column, `1` per-row, `None` for the full table.

- The `subset` parameter applies the gradient to specific columns only.
- The `vmin` and `vmax` parameters set the boundaries of the range used for color scaling.
- By default, `vmin` is the smallest value and `vmax` is the largest value.

- The `text_gradient` method applies the color gradient to the text instead of the background.

## In-Cell Bar Charts

- The `bar` method draws horizontal bars inside cells, proportional to each value.
- The `color` parameter sets the bar color.
- The `vmin` and `vmax` parameters control the bar scale.

- The `subset` parameter restricts bars to specific columns.
