## Filter A DataFrame  Based On A Condition
- Pandas needs a Boolean `Series` to perform a filter.
- Pass the Boolean `Series` inside square brackets after the `DataFrame`.

- Assign the Boolean `Series` to a variable for a simpler syntax.

- An existing Boolean column can serve as a filter.

- We can generate a Boolean `Series` using a wide variety of operations (equality, inequality, less than, greater than, inclusion, etc)

## Filter with More than One Condition (AND)

- Add the `&` operator in between two Boolean `Series` to filter by multiple conditions.
- Add parentheses around the individual `Series` to avoid errors/confusion.

- We can assign the `Series` to variables to make the syntax more readable.
- Variables also remove the need for parentheses.

## Filter with More than One Condition (OR)
- Use the `|` operator in between two Boolean `Series` to filter by *either* condition.

- Let's target employees who are either senior management OR who started before 1990-01-01.

- Let's look for employees named Robert who work in Client Services OR have a start date after 2016-06-01.

## The isin Method
- The `isin` method accepts a collection object like a list, tuple, or `Series`.
- The method returns True for a row if its value is found within the collection.

- Let's say we're targeting employees whose team is Legal, Sales, or Product.

## The isnull and notnull Methods
- The `isnull` method returns True for `NaN` values in a `Series`.
- The `notnull` method returns True for present values in a `Series`.
- The `isna` method is an alias for `isnull`.
- The `notna` method is an alias for `notnull`.

## The between Method
- The `between` method returns True if a `Series` value is found within its range.

## The duplicated Method
- The `duplicated` method returns True if a `Series` value is a duplicate.
- Pandas will mark one occurrence of a repeated value as a non-duplicate.
- Use the `keep` parameter to designate whether to consider the first or last occurrence of a repeated value as the non-duplicate.
- Pass `False` to the `keep` parameter to mark all occurrences of repeated values as duplicates.

- Use the tilde symbol (`~`) to invert a `Series's` values. True will become Falses, and Falses will become Trues.

## The drop_duplicates Method
- The `drop_duplicates` method deletes rows with duplicate values.
- By default, it will remove a row if *all* of its values are shared with another row.
- The `subset` parameter configures the columns to look for duplicate values within.

- Pass a list to `subset` parameter to look for duplicates across multiple columns.

## The where Method
- Like the square brackets, the `where` method accepts a Boolean `Series` to filter the `DataFrame`.
- The resulting `DataFrame` will be of the same length as the original.
- Rows that do not satisfy the filtering criteria will have `NaN`s.
- The `where` method is ideal when you want to preserve the original `DataFrame` structure.

## The query Method
- The `query` method filters a `DataFrame` using a string argument.
- Its argument is a more readable alternative to Boolean filtering.

- Use `and`, `or`, and `not` (or `&`, `|`, `~`) within the expression string.

- Use the `in` and `not in` operators to check inclusion/exclusion.

- Use the `@` prefix to reference local variables.
